```python
import os
import sys
import pandas as pd

# Allow notebook to access src/
sys.path.append(os.path.abspath(".."))

from src.data_loader import extract_text


```

```python
BASE_PATH = "../data/raw/Resumes"
```

```python
data = []

for category in os.listdir(BASE_PATH):
    category_path = os.path.join(BASE_PATH, category)

    # Only process folders (categories)
    if os.path.isdir(category_path):
        for file in os.listdir(category_path):
            if file.lower().endswith((".pdf", ".docx", ".doc")):
                file_path = os.path.join(category_path, file)

                text = extract_text(file_path)

                # Keep only valid resumes
                if text:
                    data.append({
                        "resume_text": text,
                        "category": category,
                        "file_name": file
                    })

for category in os.listdir(BASE_PATH):
    print("Found category folder:", category)
```

```python
df = pd.DataFrame(data)
```
