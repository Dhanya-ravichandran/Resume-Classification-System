```python
import os
import sys
import pandas as pd

sys.path.append(os.path.abspath(".."))

from src.nlp.vectorization import get_tfidf_vectorizer
```

```python
df = pd.read_csv("../data/processed/resumes_cleaned.csv")

X_text = df["clean_text"]
y = df["category"]
```

```python
vectorizer = get_tfidf_vectorizer(
    max_features=3000,
    ngram_range=(1, 2)
)
```

```python
X = vectorizer.fit_transform(X_text)
```

```python
X.shape
```

```python
import pickle
import os

os.makedirs("../models", exist_ok=True)

with open("../models/tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

print(" Fitted TF-IDF vectorizer saved")
```

```python
import pickle

with open("../models/tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

with open("../models/X_tfidf.pkl", "wb") as f:
    pickle.dump(X, f)

with open("../models/y_labels.pkl", "wb") as f:
    pickle.dump(y, f)

print("TF-IDF features and vectorizer saved")
```

TF-IDF vectorization was applied to convert cleaned resume text into numerical feature vectors by capturing both term importance and contextual word patterns, enabling effective supervised classification.
